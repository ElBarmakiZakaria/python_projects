```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from pathlib import Path



np.random.seed(42) # make this notebook's output stable across runs

# Let's define the default font sizes, to plot pretty figures
plt.rc('font', size=12)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=12)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)


# Define pandas library settings
pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
# pd.set_option('display.mac_colwidth', None)
# pd.set_option('display.precision', 2)
# pd.reset_option('display.max_columns')
```

```python
datapath = Path() / "data" / "lifesat"

oecd_bli = pd.read_csv(datapath / "oecd_bli.csv")
gdp_per_capita = pd.read_csv(datapath / "gdp_per_capita.csv")
```

```python
oecd_bli.head(10)
```

```python
gdp_per_capita
```

```python
gdp_per_capita_2020 = gdp_per_capita[gdp_per_capita["Year"]==2020]
gdp_per_capita_2020 = gdp_per_capita_2020.rename(
    columns={"GDP per capita, PPP (constant 2017 international $)": "GDP per capita (USD)"}
)
gdp_per_capita_2020
```

```python
# life satisfaction is one of the values in 'indicator' column
oecd_bli["Indicator"].value_counts()
```

```python
# Preprocess the OECD BLI data to keep only the life satisfaction column
# We need to create a pivot table with appropriate values
oecd_bli = oecd_bli[oecd_bli["INEQUALITY"]=="TOT"]
oecd_bli = oecd_bli.pivot(index="Country", columns="Indicator", values="Value")
oecd_bli
```

```python
oecd_bli = oecd_bli.reset_index()
oecd_bli
```

```python
# Display dataframe info
gdp_per_capita_2020.info()
```

```python
# Now let's merge the lige satisfaction data and the GDP per capita data
# "Country" and "Entity" columns represent the same values
full_country_stats = oecd_bli.merge(
    gdp_per_capita_2020, left_on="Country", right_on="Entity", how="inner"
)
full_country_stats
```

```python
# lets keep only Country, GDP per capita and life satisfaction columns and then sort data
full_country_stats = full_country_stats[["Country", "Life satisfaction", "GDP per capita (USD)"]]
full_country_stats = full_country_stats.sort_values(by="GDP per capita (USD)")
full_country_stats
```

```python
full_country_stats.to_csv(datapath / "lifesat_full.csv", index=False)
```

# chapter 1

```python
min_gdp = 23_500
max_gdp = 62_500

country_stats = full_country_stats[(full_country_stats["GDP per capita (USD)"] >= min_gdp) &
                                   (full_country_stats["GDP per capita (USD)"] <= max_gdp)]
country_stats.to_csv(datapath / "lifesat.csv", index=False)
```

```python
lifesat = pd.read_csv(datapath / "lifesat.csv")
lifesat
```

```python
lifesat.plot(kind="scatter", grid=True,
             x="GDP per capita (USD)", y='Life satisfaction')

plt.show()
```

```python
# prepare model atributes 
X = lifesat[["GDP per capita (USD)"]]
y = lifesat[["Life satisfaction"]]

# sELECT linear model
model = LinearRegression()

# Train the model 
model.fit(X, y)

# Trained model parameters
theta_0 = model.intercept_[0]
theta_1 = model.coef_[0][0]
print(f"theta_0: {theta_0}")
print(f"theta_1: {theta_1}")



```

$$\frac{1}{2}\sum_{i=1}^{n}$$

```python
lifesat.plot(kind="scatter", grid=True,
             x="GDP per capita (USD)", y='Life satisfaction')

plt.plot(X, theta_0+theta_1*X)
plt.text(45000, 6.5, fr"$\theta_0 = {theta_0.round(2)}$", color="blue")
plt.text(45000, 6.3, fr"$\theta_1 = {theta_1 * 1e5:.2f}\cdot 10^{{-5}}$", color="blue")
plt.title("Relationship between GDP per capta and Life Satisfaction")
plt.savefig("gdp_per_capita_vs_life_satisfaction", dpi=200)
plt.show()
```

```python
cyprus_gdp_per_capita_2020 = (
    gdp_per_capita_2020[gdp_per_capita_2020["Entity"]=="Cyprus"][["GDP per capita (USD)"]]
)
cyprus_gdp_per_capita_2020
```

```python
cyprus_predicted_life_satisfaction = model.predict(cyprus_gdp_per_capita_2020)
cyprus_predicted_life_satisfaction
```

```python
from sklearn.neighbors import KNeighborsRegressor
model_knn = KNeighborsRegressor(n_neighbors=3)

model_knn.fit(X, y)


model_knn.predict(cyprus_gdp_per_capita_2020)
```
